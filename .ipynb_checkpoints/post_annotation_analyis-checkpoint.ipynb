```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import sklearn as skl
```

# Loading annotated files

Loading files, and inspecting relative frequencies of annotations

```python
annotated_1 = pd.read_csv('Annotation/annotation_1.csv',index_col='Unnamed: 0')
annotated_1 = annotated_1['stance']
annotated_1[annotated_1.map(lambda x: x not in ['+','.','-'])] = np.nan
annotated_1.value_counts()/sum(annotated_1.value_counts())
```

```python
annotated_2 = pd.read_csv('Annotation/annotation_2.csv',index_col='Unnamed: 0')
annotated_2 = annotated_2['stance']
annotated_2.value_counts()/sum(annotated_2.value_counts())
```

```python
annotated_3 = pd.read_csv('Annotation/annotation_3.csv',index_col='Unnamed: 0')
annotated_3 = annotated_3['stance']
annotated_3[annotated_3=='+ '] = '+'
annotated_3[annotated_3.map(lambda x: x not in ['+','.','-'])] = np.nan
annotated_3.value_counts()/sum(annotated_3.value_counts())
```

**Observation:** The third annotators has many more neutral annotations.

```python
annotated_combined = annotated_1.to_frame().join(annotated_2,lsuffix="_matthijs",rsuffix='_other')
annotated_combined.loc[annotated_3.notnull(),'stance_other'] = annotated_3
annotated_combined
two_annotators_indices = annotated_combined.dropna(thresh=2).index
```

```python
annotated_combined['stance_other'][two_annotators_indices]
```

Calculating Cohen's Kappa between me and the other annotators:

```python
from sklearn import metrics
print('Between me and the first annotator:')
print(metrics.cohen_kappa_score(list(annotated_combined['stance_matthijs'][annotated_2.notnull()]),
                          list(annotated_combined['stance_other'][annotated_2.notnull()])))
print('Between me and the second annotator:')
print(metrics.cohen_kappa_score(list(annotated_combined['stance_matthijs'][annotated_3.notnull()]),
                          list(annotated_combined['stance_other'][annotated_3.notnull()])))
print('Between me and both annotators:')
print(metrics.cohen_kappa_score(list(annotated_combined['stance_matthijs'][two_annotators_indices]),
                          list(annotated_combined['stance_other'][two_annotators_indices])))
```

Combining annotations:

```python
annotated_combined['combined'] = np.nan
stance_combininations = {'++':'+',
'+.':'+.',
'+-':'.',
'.+':'+.',
'..':'.',
'.-':'-.',
'-+':'.',
'-.':'-.',
'--':'-'}
annotated_combined['combined'] = (annotated_combined['stance_matthijs']+annotated_combined['stance_other']).map(stance_combininations)
```

```python
annotated_combined['combined']
```

```python
annotated_combined['combined'].value_counts()
```

# Creating new files for ties

```python
annotated_2 = pd.read_csv('Annotation/annotation_Vera.csv',index_col='Unnamed: 0')
annotated_2['stance'] = ''
annotated_2_ties = annotated_2[annotated_combined['combined'].isin(['+.','-.'])]
annotated_2_ties.to_csv('Annotation/annotation_2_ties.csv')
```

```python
annotated_3 = pd.read_csv('Annotation/annotation_Victor.csv',index_col='Unnamed: 0')
annotated_3['stance'] = ''
annotated_3_ties = annotated_3[annotated_combined['combined'].isin(['+.','-.'])]
annotated_3_ties.to_csv('Annotation/annotation_3_ties.csv')
```

# Resolving ties

```python
annotated_2_tiebreaker = pd.read_csv('Annotation/annotation_2_ties_completed.csv',index_col='Unnamed: 0')
annotated_3_tiebreaker = pd.read_csv('Annotation/annotation_3_ties_completed.csv',index_col='Unnamed: 0')
```

```python
annotated_2_tiebreaker
```

```python
annotated_combined['tiebreaker'] = ''
annotated_combined.loc[annotated_2_tiebreaker.index,'tiebreaker'] = annotated_2_tiebreaker['stance']
annotated_combined.loc[annotated_3_tiebreaker.index,'tiebreaker'] = annotated_3_tiebreaker['stance']
```

```python
annotated_combined['tiebreaker'].value_counts()
```

```python
annotated_combined['combined_tie'] = np.nan
stance_combininations = {'+':'+',
                        '.':'.',
                        '+.+':'+',
                        '-':'-',
                        '+..':'.',
                        '-.-':'-',
                        '-..':'.',
                        '+.-':'.',
                        '-.+':'.'}
annotated_combined['combined_final'] = (annotated_combined['combined']+annotated_combined['tiebreaker']).map(stance_combininations)
```

```python
annotated_combined['combined_final'].value_counts()
```

# Combining with full data

```python
full_data = pd.read_csv('Annotation/full_data.csv',index_col='Unnamed: 0').drop('stance',axis=1)
full_data = full_data.join(annotated_combined['combined_final'])
full_data = full_data.rename({'combined_final':'stance'},axis=1)
annotated_data = full_data.loc[two_annotators_indices]
```

```python
annotated_data
```

# Analysis

```python
print('Annotations per period:')
count_recent = sum(annotated_data["recent"])
count_former = sum(annotated_data["former"])
print(f'Recent: {count_recent}')
print(f'Former: {count_former}')

print()
print('Annotations in both periods:')
print(sum(annotated_data[["recent","former"]].all()))

print()
print('Off the annotated data, percentages outcome in the recent period:')
stance_count_recent = annotated_data.loc[annotated_data['recent'],'stance'].value_counts()
stance_count_former = annotated_data.loc[annotated_data['former'],'stance'].value_counts()

print(f'+: {stance_count_recent.get("+",0)/count_recent}')
print(f'.: {stance_count_recent.get(".",0)/count_recent}')
print(f'-: {stance_count_recent.get("-",0)/count_recent}')
print('Off the annotated data, percentages outcome in the former period:')
print(f'+: {stance_count_former.get("+",0)/count_former}')
print(f'.: {stance_count_former.get(".",0)/count_former}')
print(f'-: {stance_count_former.get("-",0)/count_former}')
```

### Label Proportions

```python
ax = plt.subplot(111)

labels_stance = ['negative','neutral','positive']
stances_perc_recent = [x/count_recent*100 for x in stance_count_recent.values]
stances_perc_recent.reverse()
stances_perc_former = [x/count_former*100 for x in stance_count_former.values]
stances_perc_former.reverse()

width=0.6
breaks=1.2
_X = (width*breaks)*np.arange(len(labels_stance)) 
plt.bar(_X - width/2, stances_perc_former, width=width/2, 
        align="edge",label='earlier',color='tab:olive')  
plt.bar(_X, stances_perc_recent, width=width/2, 
        align="edge",label='recent',color='tab:blue')  
plt.xticks(_X, labels_stance)
plt.ylim((0,100))
plt.ylabel('Percentage')
#plt.title('Percentage of Each Stance for Both Periods')
plt.legend()
plt.show()
plt.savefig('annotation_percentage.png', dpi=1200)
```

```python
from scipy.stats import mannwhitneyu
mannwhitneyu(annotated_data.loc[annotated_data['recent'],'stance'].map({'+':1,'.':0,'-':-1}),
             annotated_data.loc[annotated_data['former'],'stance'].map({'+':1,'.':0,'-':-1}))
```

### Exploring themes

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.min_rows',80)
annotated_data[(annotated_data['stance']=='+') & (annotated_data['former'])].sort_values(by='mean_score') \
        [['body','mean_score','nr_comments','stance','former']]
```

### Score analysis

```python
scores_recent = annotated_data.loc[annotated_data['recent'],'total_score']
scores_former = annotated_data.loc[annotated_data['former'],'total_score']
print(f'Recent mean: {scores_recent.mean()}, standard deviation: {scores_recent.std()}')
print(f'Former mean: {scores_former.mean()}, standard deviation: {scores_former.std()}')
```

```python

ax = plt.subplot(111)
stance_scores_recent = annotated_data.loc[annotated_data['recent']].groupby('stance') \
        .agg({'mean_score':['median', 'std']})['mean_score']
stance_scores_former = annotated_data.loc[annotated_data['former']].groupby('stance')\
        .agg({'mean_score':['median', 'std']})['mean_score']

labels_stance = ['negative','neutral','positive']
stance_scores_recent = stance_scores_recent.reindex(['-','.','+'])
stance_scores_former = stance_scores_former.reindex(['-','.','+'])

width=0.6
breaks=1.2
_X = (width*breaks)*np.arange(len(labels_stance)) 
plt.bar(_X - width/2, stance_scores_former['median'], width=width/2, 
        align="edge",label='earlier',color='tab:olive')  
plt.bar(_X, stance_scores_recent['median'], width=width/2, 
        align="edge",label='recent',color='tab:blue')  
plt.xticks(_X, labels_stance)
plt.ylabel('Median score')
plt.title('Median of the User Scores for the different Periods and Labels.')
plt.legend()
plt.show()
```

Running Spearman's r tests between stance and score for both periods:

```python
from scipy import stats
stats.spearmanr(annotated_data.loc[annotated_data['recent'],'stance'].map({'+':1,'.':0,'-':-1}),
                annotated_data.loc[annotated_data['recent'],'mean_score'])
```

```python
stats.spearmanr(annotated_data.loc[annotated_data['former'],'stance'].map({'+':1,'.':0,'-':-1}),
                annotated_data.loc[annotated_data['former'],'mean_score'])
```
